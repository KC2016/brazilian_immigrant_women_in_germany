```python
df = df.replace({'reason_to_migrate' : {'Cônjuge/namorado(a)' : 'spouse_partner', 'Busca de novas oportunidades' : 'search_for_new_opportunities', 'Família (pais, filhos)' : 'family', 'Trabalho': 'job'}})
# df.head()
default_reason_to_migrate = ['spouse_partner', 'search_for_new_opportunities', 'family', 'job']
# df.head()]

filtered_df = df[~df['reason_to_migrate'].isin(default_reason_to_migrate)]

# Get unique values in the 'reason_to_migrate' column after filtering
unique_reasons = filtered_df['reason_to_migrate'].unique()
unique_reasons
```

```python
studies_related = ['mestrado', 'Estudo', 'Intercâmbio ', 'Estudo (doutorado)', 'Estudar o idioma', \
                   'Aprender alemão, morar com irmã ', 'Fazer mestrado ', 'Estudar ', 'Estudos', 'Estudos ', 'Mestrado ']
job_related = ['Au pair', 'Trabalho ']
spouse_partner_related = ['Trabalho do meu marido',  'era nova e tinha muitas ilusoes. \
O amor, uma carreira, a Europa que eu costumo chamar brincando de "Zorópias"... e tinha uma pitada grande de viralatice na minha essência pessoal.\
 Cá estou há 24 anos... O amor continua, mas já o resto... se eu soubesse que seriam tantos desafios tinha ficado quietinha no BR.']

other_related = ['Tínhamos vontade de deixar o Brasil', \
                 'Ter a experiência de morar fora do Brasil', \
                 'Qualidade de vida e mais segurança', \
                 'Conhecer outra cultura',\
                 'Na primeira vez foi acompanhando meu marido no PHD dele, e após voltarmos ao Brasil e vivermos alguns anos lá a vida acabou nos trazendo de volta e desta vez para ficar.']


# df.head()

df['reason_to_migrate'] = df['reason_to_migrate'].str.strip()

for v in df['reason_to_migrate']:
  if v in studies_related:
    df['reason_to_migrate']=df['reason_to_migrate'].replace(v, 'studies')
  elif v in job_related:
    df['reason_to_migrate']=df['reason_to_migrate'].replace(v, 'job')
  elif v in spouse_partner_related:
    df['reason_to_migrate']=df['reason_to_migrate'].replace(v, 'spouse_partner')
  elif v in other_related:
    df['reason_to_migrate']=df['reason_to_migrate'].replace(v, 'other')

df['reason_to_migrate']=df['reason_to_migrate'].replace('Cônjuge e busca de novas oportunidades', 'spouse_partner')


for v in df['reason_to_migrate'].unique():
  print(v)
```
